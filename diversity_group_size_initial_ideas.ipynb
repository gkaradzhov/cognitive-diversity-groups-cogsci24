```python
import ast
```

```python
a = ast.literal_eval("Counter({'OTHER_ERROR': 3, 'BIASED': 2})"[8:-1])
```

```python
from collections import defaultdict
```

```python
biased = []
diverse = []
contain_correct = []
contain_correct_and_diverse = []

for _, item in df.iterrows():
    solo_submissions_fine = ast.literal_eval(item['onboarding_types_fine'][8:-1])
    solo_submissions =  ast.literal_eval(item['onboarding_types'][8:-1])
    df.at[_, 'diversity'] = len(solo_submissions_fine)

    if item['num_of_players'] == 2:
        continue
    if len(solo_submissions_fine) <= 2 and 'CORRECT' not in solo_submissions:
        biased.append(item['performance_gain'])
    if len(solo_submissions_fine) >=4 and 'CORRECT' not in solo_submissions:
        diverse.append(item['performance_gain'])
    if 'CORRECT' in solo_submissions and len(solo_submissions_fine) <= 2:
        contain_correct.append(item['performance_gain'])
        
    if 'CORRECT' in solo_submissions and len(solo_submissions_fine) >=4:
        contain_correct_and_diverse.append(item['performance_gain'])

print(len(diverse))
print(len(contain_correct_and_diverse))
```

```python

all_pgs = []
for item in contain_correct_and_diverse:
    all_pgs.append([1, 1, item])
for item in contain_correct:
    all_pgs.append([0, 1, item])
for item in diverse:
    all_pgs.append([1, 0, item])
for item in biased:
    all_pgs.append([0, 0, item])
```

```python
len(all_pgs)
```

```python
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm
```

```python
df_new = pd.DataFrame(all_pgs, columns=['diverse', 'correct', 'pg'])
df_new.describe()
```

```python
from scipy.stats import zscore
```

```python
df_new[['diverse']] = df_new[['diverse']].transform(lambda x: zscore(x))
df_new[['correct']] = df_new[['correct']].transform(lambda x: zscore(x))

df_new[['pg']] = df_new[['pg']].transform(lambda x: zscore(x))
```

```python
all_trans = ols("pg ~ diverse  + correct", data=df_new).fit()
all_trans.summary()
```
